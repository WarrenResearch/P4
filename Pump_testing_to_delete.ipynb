```python
from jasco_pump232 import JascoPU2080
import serial

```

```python
com_port = 4

try:
    pump = JascoPU2080(com_port)
    print(f"Connected to pump on COM{com_port}")

    # Test communication with a basic command that should always respond
    # '0 id load p' often works; some pumps may respond to '0 ver load p' or '0 pump load p'
    try:
        response = pump._comm("read", "0 id load p")
        print("Pump responded:", response)
    except RuntimeError:
        print("No response from pump. Check serial settings, COM port, and pump address.")

finally:
    pump.close()
    print("Connection closed.")
```

```python
try:
    pump.is_running()
finally:
    pump.close()
```

```python
for baud in [2400, 4800, 9600]:
    try:
        ser = serial.Serial("COMx", baudrate=baud, timeout=1, rtscts=True)
        ser.write(b"a_flow load p\r")
        print(baud, ser.readline())
        ser.close()
    except Exception as e:
        print(baud, e)
```
